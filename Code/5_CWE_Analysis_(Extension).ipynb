```python
# Display the 15 most common values in the 'cwe' column
android_common_cwe = df_android['cwe'].value_counts().head(15)

# Retrieve and print the cwe_info for the most common CWEs
for cwe, count in android_common_cwe.items():
    cwe_info = df_android[df_android['cwe'] == cwe]['cwe_info'].iloc[0]
    print(f"{cwe:15} | Count: {count:6} | Info: {cwe_info:}")
```

```python
print("\nSeverity Distribution for Common CWEs:\n")
# Retrieve and print the severity distribution for the most common CWEs
for cwe, count in android_common_cwe.items():
    severity = df_android[df_android['cwe'] == cwe]['base_severity'].value_counts().head(15)
    print(f"{cwe}:")
    for s, c in severity.items():
        print(f"Severity: {s:10} | Count: {c}")
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import kaleido

severity_data = []

for cwe in df_android['cwe'].unique():
    cwe_counts = df_android[df_android['cwe'] == cwe].shape[0]
    severity_data.append({'cwe': cwe, 'count': cwe_counts})

severity_df = pd.DataFrame(severity_data)

# Calculate total counts per CWE
cwe_counts = severity_df.groupby('cwe')['count'].sum().reset_index()
cwe_counts = cwe_counts.sort_values(by='count', ascending=False)

fig = px.treemap(cwe_counts, path=['cwe'], values='count')
fig.update_traces(textinfo='label+value')

# Center the title and adjust the size
fig.update_layout(
    title_x=0.5,
    font=dict(size=14),
    width=1200,  # Set the width of the figure
    height=400  # Set the height of the figure
)
fig.write_image('../2024_Data/Figures/cwe_treemap.png')

fig.show()
```

```python
print("\nSeverity Distribution for Common CWEs:\n")
# Retrieve and print the severity distribution for the most common CWEs
for cwe, count in huawei_common_cwe.items():
    severity = df_huawei[df_huawei['cwe'] == cwe]['base_severity'].value_counts().head(15)
    print(f"{cwe}:")
    for s, c in severity.items():
        print(f"Severity: {s:10} | Count: {c}")
    print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

severity_data = []

for cwe, count in huawei_common_cwe.items():
    severity_counts = df_huawei[df_huawei['cwe'] == cwe]['base_severity'].value_counts().reset_index()
    severity_counts.columns = ['severity', 'count']
    severity_counts['cwe'] = cwe
    severity_data.append(severity_counts)

severity_df = pd.concat(severity_data, ignore_index=True)

plt.figure(figsize=(20, 10))
sns.barplot(x='cwe', y='count', hue='severity', data=severity_df)
plt.title('Severity Distribution for Top 15 Most Common CWEs in Huawei Security Bulletins', fontsize=20)
plt.xlabel('CWE', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.legend(title='Severity', loc='upper right', title_fontsize=18, fontsize=16)
plt.tight_layout()

# Extract cwe_info for CWEs in the chart
unique_cwes = huawei_common_cwe.index
cwe_info_texts = [f"{cwe}: {df_huawei[df_huawei['cwe'] == cwe]['cwe_info'].iloc[0]}" for cwe in unique_cwes]
cwe_info_text = "\n".join(cwe_info_texts)

# Add the cwe_info text at the bottom of the plot
plt.gcf().text(0.02, -0.15, cwe_info_text, fontsize=16, ha='left', va='top')

# Save the plot as an image file
plt.savefig("../2024_Data/Figures/huawei_severity_distribution.png", bbox_inches="tight")

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to get top 15 common CWEs and their severity counts
def get_top_cwes_with_severity(df, source):
    common_cwe = df['cwe'].value_counts().head(15)
    common_cwe_df = common_cwe.reset_index()
    common_cwe_df.columns = ['cwe', 'count']
    common_cwe_df['source'] = source
    common_cwe_df['cwe_info'] = common_cwe_df['cwe'].apply(lambda x: df[df['cwe'] == x]['cwe_info'].iloc[0] if not df[df['cwe'] == x]['cwe_info'].isna().empty else '')

    # Calculate severity counts
    severity_data = []
    for cwe in common_cwe_df['cwe']:
        severity_counts = df[df['cwe'] == cwe]['base_severity'].value_counts().reset_index()
        severity_counts.columns = ['severity', 'count']
        severity_counts['cwe'] = cwe
        severity_data.append(severity_counts)
    
    severity_df = pd.concat(severity_data, ignore_index=True)
    severity_df['source'] = source

    return severity_df, common_cwe_df

# Define a function to plot top CWEs with severity for a given dataset
def plot_cwes_with_severity(severity_df, ax):
    sns.barplot(x='cwe', y='count', hue='severity', data=severity_df, ax=ax)
    ax.set_title(f'Most Common CWEs - {severity_df["source"].iloc[0]}', fontsize=16)
    ax.set_xlabel('CWEs', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.tick_params(axis='x', rotation=75)
    ax.legend(title='Severity', loc='upper right', title_fontsize=12, fontsize=10)

# Get top 15 CWEs with severity for each dataset and collect cwe_info
df_dict = {
    'Android Security Bulletin': df_android,
    'Huawei': df_huawei,
    'Vivo': df_vivo,
    'Pixel': df_pixel,
    'Samsung': df_samsung,
    'Motorola': df_motorola,
    'LG': df_lg,
    'Oppo': df_oppo,
    'Xiaomi': df_xiaomi
}

all_common_cwe = pd.DataFrame()

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))
axes = axes.flatten()

# Plot each dataset and combine common CWEs
for ax, (source, df) in zip(axes, df_dict.items()):
    severity_df, common_cwe_df = get_top_cwes_with_severity(df, source)
    plot_cwes_with_severity(severity_df, ax)
    all_common_cwe = pd.concat([all_common_cwe, common_cwe_df], ignore_index=True)

plt.tight_layout()

# Aggregate counts for each CWE across all OEMs
cwe_summary = all_common_cwe.groupby('cwe').agg({'count': 'sum', 'cwe_info': 'first'}).reset_index()
cwe_summary = cwe_summary.sort_values(by='count', ascending=False)
print(cwe_summary)
print()
print()
print()
print()


# # Format the cwe_info text
# cwe_info_texts = [f"{count:10} {cwe:20}: {info}" for cwe, count, info in zip(cwe_summary['cwe'], cwe_summary['count'],cwe_summary['cwe_info'])]
# cwe_info_text = "\n".join(cwe_info_texts)

# # Add the cwe_info text at the bottom of the plot
# plt.gcf().text(0.02, -0.15, cwe_info_text, fontsize=16, ha='left', va='top')

# Save the plot as an image file
plt.savefig("../2024_Data/Figures/common_cwe_severity_across_venders.png", bbox_inches="tight")

# Display the plot
plt.show()
```
